```python
import re
from collections import Counter
import pandas as pd

```

```python
def read_data(file):
    data = []
    lab=[]
    l=[]
    with open(file, 'r')as f:
        
        for line in f:
            line = line.strip()
            text = line[line.find("]")+1:].strip()
            
            label = ' '.join(line[1:line.find("]")].strip().split())
            label=re.sub('[.]',",",label)
            label=re.sub('[ ]','',label)
            label=re.sub('[,]','',label)
            
            lab=[int(i)  for  i in label]
           
            l.append(lab)
            data.append([text])
           
            
    return data,l
file='/content/drive/MyDrive/Colab Notebooks/Datasets/emotion.txt'
text,labels= read_data(file)
```

```python

def get_emotions(label):
  lab=[]
  emotions = ["joy", 'fear', "anger", "sadness", "disgust", "shame", "guilt"]
  for j in label:
    
    for i in range(len(j)):
      if j[i]==1:
        lab.append([emotions[i]])
        
        break
  return lab 
final_labels=get_emotions(labels)


```

```python


data=[]
for i in range(len(text)):
  data.append([final_labels[i],text[i]])
```

```python
x=text 
y=final_labels
import numpy as np
df=pd.DataFrame(final_labels)
df.rename(columns={0:'Emotions'},inplace=True)
df["Text"]=pd.DataFrame(text)
print(df.head())
```

```python
x=df.iloc[:,1].values
y=df.iloc[:,0].values
print(y.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000)
x=cv.fit_transform(x).toarray()
```

```python
print(x[0])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.2,random_state=42)
```

```python

from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
lsv=LinearSVC(random_state=42)
dtc=DecisionTreeClassifier(random_state=42)
rfc=RandomForestClassifier()
nv=MultinomialNB()
sv=SVC()
cls=[lsv,dtc,rfc,nv,sv]
```

```python
def check_best_methods(clf,x_train,x_test,y_train,y_test):
  clf.fit(x_train,y_train)
  y_pred=clf.predict(x_test)
  test_acc=accuracy_score(y_test,y_pred)
 
  print("test_accuracy for {} :- {} ".format(clf,test_acc))
  return

for i in cls:
  check_best_methods(i,x_train,x_test,y_train,y_test)
```
